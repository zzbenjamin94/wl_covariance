## Computation of $\Delta\Sigma$ for the most massive clusters

```python
from tqdm import tqdm
##Function to find halos within vicinity
dr = 0.1
dz = 100

def deltasigma_cluster(x, y, z, r_range, dz=100 ):
    global halo_df

    delta_sigma = np.zeros(len(r_range)-1)
    for i in tqdm(range(len(r_range)-1)):
        dr = r_range[i+1] - r_range[i]
        r_cur = r_range[i]
        
        ##Find halos in the annulus range. 
        mask = (halo_df.x-x)**2 + (halo_df.y-y)**2. <= (r_cur+dr)**2
        mask &= np.abs(halo_df.z - z) <= dz
        annulus_df = halo_df[mask]
        
        #Find 2D density within the annulus
        area = np.pi * ((r_cur+dr)**2 - (r_cur)**2.)
        mvir_tot = annulus_df.mvir.sum()
        delta_sigma[i] = mvir_tot/area
        
    return delta_sigma
    
```

```python
r_range = np.linspace(0,10,100)
delta_sigma = deltasigma_cluster(949.97443, 874.50427, 257.60870, r_range, dz=200)
```

```python
## First fix the problem with matplotlib. 

plt.figure(figsize=(12,4))
dz = [50,100,200]
r_range = np.logspace(0,1,100)
delta_sigma_base = deltasigma_cluster(949.97443, 874.50427, 257.60870, r_range, dz=50)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,4))
for i, dz_cur in enumerate(dz):
    delta_sigma = deltasigma_cluster(949.97443, 874.50427, 257.60870, r_range, dz=dz_cur)

    ax1.plot(r_range[:-1], delta_sigma, label='dz={:d}'.format(dz_cur))
    ax1.set_yscale('log'); ax1.set_xscale('log')
    ax1.set_ylabel(r'$\Delta\Sigma$')
    ax1.set_xlabel('R (Mpc)')
             
    ax2.plot(r_range[:-1], delta_sigma/delta_sigma_base, label='dz={:d}'.format(dz_cur))
    ax2.set_xlabel('R (Mpc)')
    ax2.set_xscale('log')

plt.legend()
plt.show()
```

### Computation of DeltaSigma for periodic boundary conditions. 

```python
from tqdm import tqdm
##Function to find halos within vicinity
dr = 0.1
dz = 100

def deltasigma_cluster_periodic(x, y, z, r_range, dz=100, box_length=1000):
    global halo_df

    delta_sigma = np.zeros(len(r_range)-1)
    for i in tqdm(range(len(r_range)-1)):
        dr = r_range[i+1] - r_range[i]
        r_cur = r_range[i]
        
        
        ##Periodic boundary condition for annulus. Halos are conditions at [0,1000] Mpc boundaries.       
        dx_sqr = np.asarray([(halo_df.x-x)**2, (halo_df.x-x+box_length)**2, (halo_df.x-x-box_length)**2]).min(0)
        dy_sqr = np.asarray([(halo_df.y-y)**2, (halo_df.y-y+box_length)**2, (halo_df.y-y-box_length)**2]).min(0)
        dz_min = np.asarray([np.abs(halo_df.z-z), np.abs(halo_df.z-z+box_length), np.abs(halo_df.z-z-box_length)]).min(0)
        mask = dx_sqr + dy_sqr <= (r_cur+dr)**2
        mask &= dz_min < dz
        
        #mask = (halo_df.x-x)**2 + (halo_df.y-y)**2. <= (r_cur+dr)**2
        #mask &= np.abs(halo_df.z - z) <= dz
        
        annulus_df = halo_df[mask]
        
        #Find 2D density within the annulus
        area = np.pi * ((r_cur+dr)**2 - (r_cur)**2.)
        mvir_tot = annulus_df.mvir.sum()
        delta_sigma[i] = mvir_tot/area
        
    return delta_sigma
```

```python
r_range = np.logspace(-1,1,100)
delta_sigma_periodic = deltasigma_cluster_periodic(949.97443, 874.50427, 257.60870, r_range, dz=200)
```

```python
plt.plot(r_range[:-1], delta_sigma_periodic, label='periodic')
plt.plot(r_range[:-1], delta_sigma, label='hard cut')
plt.xscale('log'); plt.yscale('log')
plt.legend()
plt.show()
```

## Periodic boundary conditions.
Move along the z-axis (or any arbitrary axis) and show that the density inside a moving cylinder is continuous and periodic. 

```python
from tqdm import tqdm
##Function to find halos within vicinity
dl = 10
dz = 10

def local_density(x, y, z, dl=10, dz=10):
    global halo_df
        
    ##Find halos inside a box 
    mask = (np.abs(halo_df.x-x) <= dl) & (np.abs(halo_df.y-y) <= dl)
    mask &= np.abs(halo_df.z - z) <= dz
    local_df = halo_df[mask]
        
    #Find 2D density within the annulus
    area = dl**2 * dz
    mvir_tot = local_df.mvir.sum()
    rho_local = mvir_tot/area
    return rho_local
```

```python
# Make a moving cylinder and show that the density along the cylinder is moving with it. 
x = [920,950,980]; y=[850,900,950]; z_range = np.linspace(0,1000,200)

density_z = np.zeros((len(x),len(y),len(z_range)))
for i in range(len(x)):
    for j in range(len(y)):
        print(i,j)
        for k, z_cur in enumerate(z_range):
            density_z[i,j,k] = local_density(x[i], y[j], z_cur, dl=10, dz=10)
```

```python
## Extend this analysis to all halos in the box. 

for i in range(len(x)):
    for j in range(len(y)):
        plt.figure(figsize=(12,16))
        plt.subplot(3,3,i+j+1) 
        plt.plot(z_range, density_z[i,j], label="(x,y)=({:d},{:d})".format(x[i],y[i]))
        plt.plot(z_range+1000, density_z[i,j])
    
        plt.legend()
        plt.show()
```

```python
## Compare the error with the error between steps. Autocorrelation function. 
from statsmodels.graphics import tsaplots
import matplotlib.pyplot as plt

#plot autocorrelation function
fig = tsaplots.plot_acf(density_z[0,0], lags=10)
plt.show()
```
