## Computing DS from the cluster class objects. 

```python
sparse_sample = 1000
file_num = list(range(25))

'''
Support function to read the particle file. It is an (N,3) array specifiying positions of DM particles, sparsely sampled nth=1000, with mass resolution  of 1.5e9 Msun/h.
'''
def read_particle(filename, nth=1000):
    with open(filename, 'rb') as handle:
        ptcl_pos = pickle.load(handle)
    return ptcl_pos


def read_halo_file(clusters_dir, num):
    file_name = 'mdpl2_rockstar_z0p00_{}.pickle'.format(num)
    with open(clusters_dir+file_name, 'rb') as handle:
        halo_df = pickle.load(handle)
    return halo_df


class clusters(object):
    global ptcl_mass
    global ptcl
    
    #Input the cluster dataframe  
    def __init__(self, ptcl_filename, cl_df, r_range_norm, verbose=True):
        assert isinstance(cl_df, pd.core.frame.DataFrame), "cluster needs to be a dataframe"
        
        if verbose:
            print("Number of clusters in is:", len(cl_df))
            print("cluster columns: \n", list(cl_df.columns))

        self.cluster_df = cl_df
        self.DS_dz50 = np.zeros((len(cl_df), len(r_range)-1 ))  
        self.DS_dz100 = np.zeros((len(cl_df), len(r_range)-1 ))  
        self.DS_dz200 = np.zeros((len(cl_df), len(r_range)-1 ))  
        self.DS_dz200_halo = np.zeros((len(cl_df), len(r_range)-1 ))  
        
        self.r_range = np.array(r_range_norm*cl_df.rvir[:,np.newaxis])/1000 #Convert from kpc to Mpc.
        
        self.mass_dz50 = np.zeros(len(cl_df))
        self.mass_dz100 = np.zeros(len(cl_df))
        self.mass_dz200 = np.zeros(len(cl_df))
        self.mass_dz200_halo = np.zeros(len(cl_df))
        
    def __deltasigma_cluster_periodic(self, ptcl, cluster_ind, x, y, z, r_range, dz=100, box_length=1000):
        delta_sigma = np.zeros(len(r_range_norm)-1)
    
        ##Periodic boundary condition for annulus. Halos are conditions at [0,1000] Mpc boundaries.       
        dx_sqr = np.asarray([(ptcl[:,0]-x)**2, (ptcl[:,0]-x+box_length)**2, (ptcl[:,0]-x-box_length)**2]).min(0)
        dy_sqr = np.asarray([(ptcl[:,1]-y)**2, (ptcl[:,1]-y+box_length)**2, (ptcl[:,1]-y-box_length)**2]).min(0)
        dz_min = np.asarray([np.abs(ptcl[:,2]-z), np.abs(ptcl[:,2]-z+box_length), np.abs(ptcl[:,2]-z-box_length)]).min(0)

        for i in range(len(r_range)-1):
            #radius for annulus
            dr = self.r_range[cluster_ind, i+1] - self.r_range[cluster_ind,i]
            r_cur = self.r_range[cluster_ind, i]

            #Masking
            mask = dx_sqr + dy_sqr <= (r_cur+dr)**2
            mask &= dz_min < dz

            #Building an annulus
            annulus_df = ptcl[mask]

            #Find 2D density within the annulus
            area = np.pi * ((r_cur+dr)**2 - (r_cur)**2.)
            m_tot = len(annulus_df)*sparse_sample*ptcl_mass
            
            self.mass_dz50[cluster_ind] += m_tot
            self.mass_dz100[cluster_ind] += m_tot
            self.mass_dz200[cluster_ind] += m_tot
            delta_sigma[i] = m_tot/area

        return delta_sigma
    
    def __deltasigma_cluster_periodic_halo(self, halo_df, cluster_ind, x, y, z, r_range, dz=100, box_length=1000):
        delta_sigma = np.zeros(len(r_range)-1)
    
        ##Periodic boundary condition for annulus. Halos are conditions at [0,1000] Mpc boundaries.       
        dx_sqr = np.asarray([(halo_df.x-x)**2, (halo_df.x-x+box_length)**2, (halo_df.x-x-box_length)**2]).min(0)
        dy_sqr = np.asarray([(halo_df.y-y)**2, (halo_df.y-y+box_length)**2, (halo_df.y-y-box_length)**2]).min(0)
        dz_min = np.asarray([np.abs(halo_df.z-z), np.abs(halo_df.z-z+box_length), np.abs(halo_df.z-z-box_length)]).min(0)

        for i in range(len(r_range)-1):
            #radius for annulus
            dr = r_range[i+1] - r_range[i]
            r_cur = r_range[i]

            #Masking
            mask = dx_sqr + dy_sqr <= (r_cur+dr)**2
            mask &= dz_min < dz

            #Building an annulus
            annulus_df = halo_df[mask]

            #Find 2D density within the annulus
            area = np.pi * ((r_cur+dr)**2 - (r_cur)**2.)
            mvir_tot = annulus_df.mvir.sum()
            
            self.mass_dz200_halo[cluster_ind] += mvir_tot
            delta_sigma[i] = mvir_tot/area

        return delta_sigma
        
        
    def compute_DS_dz50(self):
        for i in range(len(self.cluster_df)):
            dummy = self.__deltasigma_cluster_periodic(ptcl, i, self.cluster_df.iloc[i]['x'], self.cluster_df.iloc[i]['y'], self.cluster_df.iloc[i]['z'], self.r_range_norm[i], dz=50)   
            self.DS_dz50[i] += dummy
        return
    
    def compute_DS_dz100(self):
        for i in range(len(self.cluster_df)):
            dummy = self.__deltasigma_cluster_periodic(ptcl, i, self.cluster_df.iloc[i]['x'], self.cluster_df.iloc[i]['y'], self.cluster_df.iloc[i]['z'], self.r_range_norm[i], dz=100)       
            self.DS_dz100[i] += dummy

        return
    
    def compute_DS_dz200(self):
        for i in range(len(self.cluster_df)):                
            dummy = self.__deltasigma_cluster_periodic(ptcl, i, self.cluster_df.iloc[i]['x'], self.cluster_df.iloc[i]['y'], self.cluster_df.iloc[i]['z'], self.r_range_norm[i], dz=200)    
            self.DS_dz200[i] += dummy
        return
    
    def compute_DS_dz200_halo(self):
        for i in range(len(self.cluster_df)):
            for j, num in enumerate(file_num):
                #print('file number {}'.format(num))
                
                halo_df = read_halo_file(clusters_dir, num)
               
                dummy = self.__deltasigma_cluster_periodic_halo(halo_df, i, self.cluster_df.iloc[i]['x'], self.cluster_df.iloc[i]['y'], self.cluster_df.iloc[i]['z'], self.r_range_norm[i], dz=200)
                
                self.DS_dz200_halo[i] += dummy
                                 
    def get_DS_dz50(self):
        return self.DS_dz50
    
    def get_DS_dz100(self):
        return self.DS_dz100
    
    def get_DS_dz200(self):
        return self.DS_dz200
    
    def get_DS_dz200_halo(self):
        return self.DS_dz200_halo
    
    def get_r_range(self):
        return self.r_range
    
    def get_mass_dz50(self):
        return self.mass_dz50
    
    def get_mass_dz100(self):
        return self.mass_dz100

    def get_mass_dz200(self):
        return self.mass_dz200
    
    def get_mass_dz200_halo(self):
        return self.mass_dz200_halo
```

## Building a particle based architecture

```python
from setup import ptcl_dir, cluster_dir
box_length = 1000
ptcl_mass = 1.505e9 ##Msun/h

class clusters_obj(object):

    '''
    Initializes the cluster_obj for derivation of cluster properties. 
    
    (x,y,z): The 3D coordinates of the cluster in Mpc/h
    M: Mass of the cluster in Msun/h. Could be Mvir, M200c/M200m, M500c/M500m. 
    R: The radius of the cluster in Mpc/h to normalize. Could be R200c/R200m or R500c/R500m. 
    r_range_norm: the radii range in Mpc/h over the R_vir (or R200c/R200m) for clusters
    dz: Projection depth in Mpc/h. 
    '''
    def __init__(self, cluster_id, x, y, z, M, R, Ngal, r_range_norm, dz=200):
        r_bin_center = (r_range_norm[0:-1] + r_range_norm[1:])/2
        
        self.cluster_id = cluster_id
        self.x = x; self.y = y; self.z = z
        self.M = M
        self.R = R
        self.dz = dz
        self.Sigma = np.zeros((len(x), len(r_bin_center)))     #Sigma
        self.DS = np.zeros((len(x), len(r_bin_center)))        #Delta Sigma
        self.r_range = r_range_norm*R[:,np.newaxis] #Broadcasted to (N,len(r_range))
        self.Ngal = Ngal
        return
    
    '''
    Computes the Surface Density of all the clusters. 
    '''
    def compute_Sigma_all(self):
        
        ptcl_files = glob(ptcl_dir() + 'snap_130.*')
        for file in ptcl_files:
            print(file)
            ptcl = readsnap(file, 'pos', 'dm', nth=10, suppress=1, single=True)
            
            ##Calculate the contribution of the particle file for each cluster
            ## This part of the code needs fixing. 
            
            #start = timer()
            for i in range(len(self.x)):
                #Sigma_cluster_file = self.__compute_Sigma_percluster_perfile(ptcl, self.x[i], self. y[i], self.z[i], self.r_range[i], self.dz)
                r_range = self.r_range[i]
                sigma = np.zeros(len(r_range)-1)
        
                ##Periodic boundary condition for annulus. Halos are conditions at [0,1000] Mpc boundaries.       
                dx_sqr = np.asarray([(ptcl[:,0]-x[i])**2, (ptcl[:,0]-x[i]+box_length)**2, (ptcl[:,0]-x[i]-box_length)**2]).min(0)
                dy_sqr = np.asarray([(ptcl[:,1]-y[i])**2, (ptcl[:,1]-y[i]+box_length)**2, (ptcl[:,1]-y[i]-box_length)**2]).min(0)
                dz_min = np.asarray([np.abs(ptcl[:,2]-z[i]), np.abs(ptcl[:,2]-z[i]+box_length), np.abs(ptcl[:,2]-z[i]-box_length)]).min(0)

                for j in range(len(r_range)-1):
                    #radius for annulus
                    dr = r_range[j+1] - r_range[j]
                    r_cur = r_range[j]

                    #Masking
                    mask = dx_sqr + dy_sqr <= (r_cur+dr)**2
                    mask &= dz_min < self.dz

                    #Building an annulus
                    annulus_df = ptcl[mask]

                    #Find 2D density within the annulus
                    area = np.pi * ((r_cur+dr)**2 - (r_cur)**2.)

                    m_tot = len(annulus_df)*ptcl_mass

                    #self.mass_dz50[cluster_ind] += m_tot
                    #self.mass_dz100[cluster_ind] += m_tot
                    #self.mass_dz200[cluster_ind] += m_tot
                    #delta_sigma[i] = m_tot/area

                    sigma[j] = m_tot/area
                
                self.Sigma[i] += sigma
            #end = timer()
            #print("Elapsed time: ", end-start)
            
        return
    
    '''
    Internal function to compute
    '''
    def __compute_Sigma_percluster_perfile(self, ptcl, x, y, z, r_range, dz):
        sigma = np.zeros(len(r_range)-1)
        
        ##Periodic boundary condition for annulus. Halos are conditions at [0,1000] Mpc boundaries.       
        dx_sqr = np.asarray([(ptcl[:,0]-x)**2, (ptcl[:,0]-x+box_length)**2, (ptcl[:,0]-x-box_length)**2]).min(0)
        dy_sqr = np.asarray([(ptcl[:,1]-y)**2, (ptcl[:,1]-y+box_length)**2, (ptcl[:,1]-y-box_length)**2]).min(0)
        dz_min = np.asarray([np.abs(ptcl[:,2]-z), np.abs(ptcl[:,2]-z+box_length), np.abs(ptcl[:,2]-z-box_length)]).min(0)

        for i in range(len(r_range)-1):
            #radius for annulus
            dr = r_range[i+1] - r_range[i]
            r_cur = r_range[i]

            #Masking
            mask = dx_sqr + dy_sqr <= (r_cur+dr)**2
            mask &= dz_min < dz

            #Building an annulus
            annulus_df = ptcl[mask]

            #Find 2D density within the annulus
            area = np.pi * ((r_cur+dr)**2 - (r_cur)**2.)
            m_tot = len(annulus_df)*ptcl_mass
            sigma[i] = m_tot/area
            
        return sigma
    
    def get_Sigma(self):
        return self.Sigma
    
    def get_DeltaSigma(self):
    ## DeltaSigma dimensions of M, L -- clusters and radial bins
        for i in range(len(self.x)):
            for j in range(r_bin_center):
                self.DS[i,j] = np.mean(self.Sigma[i,:j+1] - self.Sigma[i,j])
        return self.DS
    
    def get_Ngal(self):
        return self.Ngal
    
    def get_id(self):
        return self.cluster_id
    
    
```

```python
def get_DeltaSigma(Sigma, r_range_norm):
    ## DeltaSigma dimensions of M, L -- clusters and radial bins
    DS = np.zeros_like(Sigma)
    for i in range(len(Sigma)):
        for j in range(len(r_range_norm)-1):
            DS[i,j] = np.mean(Sigma[i,:j+1]) - Sigma[i,j]
    
        DS[i,0] = DS[i,1] ## replicate the inner most DS as it is 0
    
    return DS
```

```python
from kllr.kllr.regression_model import *
from kllr.kllr.regression_plotting import *


## Create function to calculate the scatter binned by radius
'''
Calculates the Covariance of the scatter of <f(X,R)> = Y in different radial bins using KLLR. 
---------------------

X: Independent variable array (e.g. Mass)
Y: profile array. Rows are the different halos and columns X-dependent profiles in each radial bins. 
Z: The global quantities (e.g Richness). Len should be number of halos. 
kernel_type: kernel for LLR
kernel_width: width of Gaussian or tophat.
'''
def calc_scatter_cov_radii(x, Y, Z, xrange=None, nbins = 10, kernel_type='gaussian', kernel_width=0.1):     
    output_Data = {}
    
    if len(x.shape) == 1: x = x[:, None] #Make sure X is atleast 2D
    xline = setup_bins(xrange, nbins, x[:, 0])
    cov_bins = np.shape(Y)[1] ## Radial bins
    correlation = np.zeros(cov_bins)
    covariance = np.zeros(cov_bins)
    
    #Within each bin apply the KLLR algorithm
    lm  = kllr_model(kernel_type=kernel_type, kernel_width=kernel_width)
    dz = lm.residuals(x, Z, bins=nbins, nBootstrap = 1,
                fast_calc = False, verbose = False, kernel_type=kernel_type, kernel_width=kernel_width)
    
    
    for i in range(cov_bins):
        #Find the data in the i/jth bin
        y_i = Y[:, i]
        
        #residuals give the scatter normalized by the mean scatter in that bin
        dy_i = lm.residuals(x, y_i, bins=nbins, nBootstrap = 1,
                  fast_calc = False, verbose = False, kernel_type=kernel_type, kernel_width=kernel_width)
           
            
        cov = np.cov(dz, dy_i) #Should consider the weights of the covariance matrix?
        covariance[i] = cov[1, 0]
        correlation[i] = cov[1, 0]/np.sqrt(cov[0,0] * cov[1,1])            
    
    output_Data['covariance'] = covariance
    output_Data['correlation'] = correlation
    return xline, output_Data
```

```python
def calc_bootstrap_cov_DS_radii(y, nBootstrap=50):     
    output_Data = {}
    
    cov_bins = len(y[0,:])
    covariance = np.zeros((cov_bins, cov_bins, nBootstrap))
    correlation = np.zeros((cov_bins, cov_bins, nBootstrap))
    
    for i in range(cov_bins):
        y_i = y[:,i]
        for j in range(cov_bins):
            y_j = y[:,j]
    
            for k in range(nBootstrap):
                #First "bootstrap" is always using unsampled data
                if k == 0:
                    rand_ind = np.ones(y_i.size).astype(bool)
                else:
                    rand_ind = np.random.randint(0, y_i.size, y_i.size)

                # Store the shuffled variables so you don't have to
                # compute the shuffle multiple times
                y_i_rand = y_i[rand_ind]
                y_j_rand = y_j[rand_ind]

                #Compute the covariance
                cov = np.cov(y_i_rand, y_j_rand)
                covariance[i,j,k] = cov[1, 0]     
                correlation[i,j,k] = cov[1, 0]/np.sqrt(cov[0,0] * cov[1,1])    
    
    output_Data['covariance'] = covariance
    output_Data['correlation'] = correlation
    return output_Data
```
