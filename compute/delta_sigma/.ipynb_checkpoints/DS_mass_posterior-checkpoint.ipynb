```python
import camb
from camb import model, initialpower

#Set cosmological parameters
pars = camb.CAMBparams()
pars.set_cosmology(H0=67.74, ombh2=0.022, omch2=0.122)
pars.set_dark_energy(w=-1.0)
pars.InitPower.set_params(ns=0.965)

#This sets the k limits and specifies redshifts
pars.set_matter_power(redshifts=[0., 0.8], kmax=2.0)

#Linear P(k)
pars.NonLinear = model.NonLinear_none
results = camb.get_results(pars)
kh, z, pk = results.get_matter_power_spectrum(minkh=1e-4, maxkh=1, npoints = 1000)

#Note: the above function has the maxkh argument for specifying a different
#kmax than was used above.
#Note: pk has the shape (N_z, N_k)

#Non-Linear spectra (Halofit)
pars.NonLinear = model.NonLinear_both
results.calc_power_spectra(pars)
khnl, znl, pknl = results.get_matter_power_spectrum(minkh=1e-4, maxkh=1, npoints = 1000)
```

```python
## Universal parameters
radii = np.logspace(-2, 3, 1000) #Mpc/h comoving
R_perp = np.logspace(-2, 2.4, 1000) #Mpc/h comoving; distance on the sky
Omega_m = 0.3089
```

```python
import cluster_toolkit
from cluster_toolkit import xi
from cluster_toolkit import deltasigma
from cluster_toolkit import bias as cluster_bias
from colossus.lss import peaks
from colossus.lss import bias
from cluster_toolkit import bias as cluster_bias

from colossus.cosmology import cosmology
from colossus.halo import concentration
cosmology.setCosmology('planck15')

def compute_DS_nonlinear(mass, concentration, halo_bias):
    global radii
    global R_perp
    global Omega_m
    global khnl, pknl
    
    xi_nfw = xi.xi_nfw_at_r(radii, mass, concentration, Omega_m)
    #Assume that k and P come from somewhere, e.g. CAMB or CLASS
    xi_mm = xi.xi_mm_at_r(radii, khnl, pknl)
    #Assume that k and P_linear came from somewhere, e.g. CAMB or CLASS
    xi_2halo = xi.xi_2halo(halo_bias, xi_mm)
    xi_hm = np.maximum(xi_nfw, xi_2halo)

    #Assume that radii and xi_hm are computed here
    Sigma = deltasigma.Sigma_at_R(R_perp, radii, xi_hm, mass, concentration, Omega_m)
    DeltaSigma_theory = deltasigma.DeltaSigma_at_R(R_perp, R_perp, Sigma, mass, concentration, Omega_m)*1e12 ## In units of Msun/Mpc^2
    
    return DeltaSigma_theory

def compute_DS_linear(mass, concentration, halo_bias):
    global Omega_m
    global kh, pk
    global radii
    global R_perp
    
    xi_mm = xi.xi_mm_at_r(radii, kh, pk)
    xi_nfw = xi.xi_nfw_at_r(radii, mass, concentration, Omega_m)
    #Assume that k and P come from somewhere, e.g. CAMB or CLASS
    xi_mm = xi.xi_mm_at_r(radii, kh, pk)
    #Assume that k and P_linear came from somewhere, e.g. CAMB or CLASS
    xi_2halo = xi.xi_2halo(halo_bias, xi_mm)
    xi_hm = np.maximum(xi_nfw, xi_2halo)

    #Assume that radii and xi_hm are computed here
    Sigma = deltasigma.Sigma_at_R(R_perp, radii, xi_hm, mass, concentration, Omega_m)
    DeltaSigma_theory = deltasigma.DeltaSigma_at_R(R_perp, R_perp, Sigma, mass, concentration, Omega_m)*1e12 ## In units of Msun/Mpc^2
    
    return DeltaSigma_theory
```

### Convert a NFW profile into a best fit mass. 
* Use the errors from the data. Relative size of the errors shouldn't matter for mass estimates. 
* Adopt the NFW using the mean mass and a nominal concentration from c-M relationship
* Get a percent level bias on the mass bias. 
* Use only the one halo term 

```python
## What is the likelihood function
'''
Gaussian likelihood with heterodescitic errors

Input:
p: 4-parameters of dimensions (4) or (4,N) with N the array of posterior samples
model: functional model for the fit (see functions below)
x: x data vector
y: y data vector
err: error on the y data vector

Returns:
lp: log probability of dimensions (N)
'''

def lnlike(p, model, x, y, err, cur_z=0):
    # the likelihood is sum of the lot of normal distributions
    #x = x[:,np.newaxis]
    #y = y[:,np.newaxis]
    #err = err[:,np.newaxis]
    
    fit = model(p,x,cur_z)
    if np.any(np.isnan(fit)):
        return -np.inf

    denom = np.power(err,2)
    lp = -0.5*np.sum(np.power((y - fit),2)/denom + np.log(denom) + np.log(2*np.pi), axis=0)
    lp = np.nan_to_num(lp, nan=-np.inf)
    return lp
```

```python
def compute_DS_onehalo(mass, c, radii_binned):
    global Omega_m
    global kh, pk
    
    xi_nfw = xi.xi_nfw_at_r(radii, mass, c, Omega_m)

    #Assume that radii and xi_hm are computed here
    Sigma = deltasigma.Sigma_at_R(R_perp, radii, xi_nfw, mass, c, Omega_m)
    DeltaSigma_onehalo = deltasigma.DeltaSigma_at_R(radii_binned, radii_binned, Sigma, mass, c, Omega_m)*1e12 ## In units of Msun/Mpc^2
    
    return DeltaSigma_onehalo
```

```python
def model_NFW(p, x, cur_z=0):
    lnM = p[0];
    c_Diemer19 = mass_concentration_Diemer19(np.exp(lnM), cur_z)
    DS_model = compute_DS_onehalo(np.exp(lnM), c_Diemer19, x)
    
    return DS_model
```

```python
'''
Calculates the DIC for a given model.
model: best fit model of dimensions (len(x),)
obs: Observed y data points
error: Error on the y-data points
'''
## Chi_squared estimator
def chisqr(model, obs, error):
    chisqr = 0
    for i in range(len(obs)):
        chisqr = chisqr + ((obs[i]-model[i])**2)/(error[i]**2)
    return chisqr
```
